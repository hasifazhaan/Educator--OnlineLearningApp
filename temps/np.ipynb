```python
import numpy  as np
import pandas as pd
from rake_nltk import Rake
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizero
from sklearn import preprocessing
```

```python
dataset = pd.read_csv("../temps/old-coursedata.csv")
```

```python
dataset.drop(2066,inplace=True)
```

```python
dataset
```

```python
Keywords = []
r = Rake()
for index, value in dataset.iterrows():
    # r.extract_keywords_from_text(row['course_title'])
    # row['Key_words'] = r.get_ranked_phrases()  
    r.extract_keywords_from_text(value['course_title'])
    Keywords.append(list(r.get_word_degrees()))
```

```python
dataset['is_paid'] = dataset['is_paid'].replace(False,0)
dataset['is_paid'] = dataset['is_paid'].replace(True,1)
```

```python
dataset['level'] = dataset['level'].replace("All Levels",0)
dataset['level'] = dataset['level'].replace("Beginner Level",1)
dataset['level'] = dataset['level'].replace("Intermediate Level",2)
dataset['level'] = dataset['level'].replace("Expert Level",3)
```

```python
convertedtime = []

for i in dataset['content_duration']:
    try:    
        data = i.split(" ")
        if data[1] == "hours":
            convertedtime.append(float(data[0])*60)
        elif data[1] =='questions':
            convertedtime.append(float(data[0])*2 )
        else:
            convertedtime.append(float(data[0]))
    except IndexError as e:
        convertedtime.append(float(data[0]))		
```

```python
dataset['convertedtime'] =  convertedtime
```

```python
subject = []
r = Rake()
for index, value in dataset.iterrows():
    # r.extract_keywords_from_text(row['course_title'])
    # row['Key_words'] = r.get_ranked_phrases()  
    r.extract_keywords_from_text(value['subject'])
    subject.append(list(r.get_word_degrees()))
```

```python
subject
```

```python
BoW = []
for i,j in zip(Keywords,subject):
    BoW.append(i+j)
Bag_Of_Words = []
for i in BoW:  
    Bag_Of_Words .append(" ".join(i))
    
```

```python
dataset['Bag_Of_Words'] = Bag_Of_Words
```

```python
dataset
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(Bag_Of_Words)
cosine_sim = cosine_similarity(count_matrix, count_matrix)
dataset['Words_Vector'] = ""
dataset['Words_Vector'] = cosine_sim
```

```python
normalized_ds = preprocessing.normalize(dataset[feature_cols])
```

```python
model = cosine_similarity(normalized_ds)
```

```python
def getCourseIndex(CourseName):
    
    # index = dataset[dataset['Bag_Of_Words'].str.contains(CourseName,regex=True)].index
    r = Rake()
    r.extract_keywords_from_text(CourseName)
    cname = " ".join(r.get_word_degrees())
    cname
    print(cname)

    course_original_name = dataset[dataset['Bag_Of_Words'].str.contains(
        cname,regex=True
        )].sort_values('Words_Vector',ascending = True)
    
    course_original_index = course_original_name.shape[0]//2
    course_original_index = course_original_name[course_original_index: course_original_index+1].index[0]
      
    return course_original_index
    
    # bool = dataset.isin([course_original_name])
    # # Getting Index Of Song If Present
    # sindex = bool[bool==True].index[0]
    # return sindex
```

```python
getCourseIndex("python")
```

```python
def recommendation(m_name,model=model):
    # default model is linear kernal 
    
    CourseIndex = getCourseIndex(m_name)
    print(CourseIndex)
    score = list(enumerate(model[CourseIndex]))
    sim_score = sorted(score,key = lambda x:x[1],reverse = True)
    sim_score = sim_score[0:11]
    
    Index = [i[0] for i in sim_score]
    print(Index)
    return dataset.loc[Index]
```

```python
recommendation("python programming")
```
